## Combine tables from other files 

```python
final_df=pd.merge(df,dfc,on='Country Code', how ='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
# Where max no of transactions occuring 
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_values=final_df.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[:3], autopct='%1.3f%%')
```

```python
fig, ax = plt.subplots()
ax.pie(country_values[:5], labels=country_names[:5])
```

India is the biggest market of zomato then USA , UK ...

```python
final_df.columns 
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
ratings.head(2)
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings,palette=['black','red','orange','yellow','green','green'])
```

Observation

    1 Not rated count is very high
    
    2 Maximum number of ratings are between 2.5 - 3.4


 CountPlot

```python
sns.countplot(x="Rating color",data=ratings, palette=['black','red','orange','yellow','green','green'])

# how frequenty its is in dataset,its not the count 
```
